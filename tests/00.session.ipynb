```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F  # noqa
from pyspark.sql import types as T  # noqa

from utils.database import MySQLManager, SQLServerManager
from utils.settings import Settings

spark = (
    SparkSession.builder.appName("")
    .config("spark.sql.caseSensitive", True)
    .config("spark.sql.session.timeZone", "UTC")
    .getOrCreate()
)
# df = spark.createDataFrame([("2022-11-12",)], ["ts"])
```

```python
config = Settings(_env_file="mysql.env")
db_manager = MySQLManager(config)
db_manager.get_jdbc_options("store")
db_manager.get_primary_key(spark, "store.TB_COMPOSITE_KEY")
db_manager.get_partition_key(spark, "store.TB_COMPOSITE_KEY")
db_manager.get_schema(spark, "store.TB_COMPOSITE_KEY")
db_manager.get_metadata(spark, "store.TB_COMPOSITE_KEY")
```

```python
config = Settings(_env_file="sqlserver.env")
db_manager = SQLServerManager(config)
db_manager.get_jdbc_options("store")
db_manager.get_primary_key(spark, "store.dbo.TB_COMPOSITE_KEY")
db_manager.get_partition_key(spark, "store.dbo.TB_COMPOSITE_KEY")
db_manager.get_schema(spark, "store.dbo.TB_COMPOSITE_KEY")
db_manager.get_metadata(spark, "store.dbo.TB_COMPOSITE_KEY")
```
